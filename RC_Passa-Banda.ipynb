### 3) Grapher Gain

```python
def graphGain(x, y, f1, Gain_cut, f2):
    
    plt.scatter(f1, Gain_cut, c = 'r' , marker = 'o') #plota a frequencia de corte
    plt.plot(x, y)#plota o gráfico inteiro 
    
    plt.xscale("log")#escala de x
    
    #plt.yticks([0, -3, -5, -7.5, -10, -12.5, -15])
    
    plt.annotate("f1", xy = (f1-180, Gain_cut+0.2)) #Anota a frequencia de corte
    plt.annotate("("+str(round(f1,2))+","+str(round(Gain_cut,3))+")",xy = (f1-220, Gain_cut-0.1),
                 fontsize="small") #anota as coordenadas
    
    plt.annotate("f2", xy = (f2+1000, Gain_cut+0.2)) #Anota a frequencia de corte
    plt.annotate("("+str(round(f2,2))+","+str(round(Gain_cut,3))+")",xy = (f2+250, Gain_cut-0.1),
                 fontsize="small") #anota as coordenadas
        
    
    plt.arrow(300, -3, 200 , 0, length_includes_head = True, linestyle = '--', color = 'g')#Banda de passagem
    plt.arrow(2200, -3, -1000 , 0, length_includes_head = True, linestyle = '--', color = 'g')
    plt.annotate("Banda\n  de\nPassagem", xy = (500, -3.5)) 
    
   
    plt.ylim(-7,0) #ajusta os eixos 
    
    #Legendas
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Ganho(dB)")
    
    #salva a figura
    plt.savefig("GraficoGanhoPassa-Banda", dpi = 400)
    
    plt.show()
```

### 6) Graph Phase

```python
def graphPhase(x, y, f_c, Ph_c):
    
    plt.scatter(f_c, Ph_c, c = 'r' , marker = 'o') #plota a frequencia de corte
    plt.plot(x, y, linestyle ="--")#plota o gráfico inteiro 
    
    plt.xscale("log")#escala de x
    
    
    plt.annotate("Frequência \n de Ressonância", xy = (f_c+100,float(Ph_c)+0.2)) #Anota a frequencia de corte
    plt.annotate("("+str(round(f_c,2))+","+str(round(float(Ph_c),3))+")",xy = (f_c+300, Ph_c-10),
                 fontsize="small") #anota as coordenadas
    
   
    
    #Legendas
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Fase(°)")
    
    #salva a figura
    plt.savefig("GraficoFasePassa-Banda", dpi = 400)
    
    plt.show()
```
